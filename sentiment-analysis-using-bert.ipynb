# Sentiment Analysis using BERT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud
```

# Reading Data


```python
df = pd.read_csv('all-data.csv', encoding='iso-8859-1')
df.head()
```

```python
df = df.rename(columns={'neutral':'sentiment','According to Gran , the company has no plans to move all production to Russia , although that is where the company is growing .':'Statement'}) # renaming the columns
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates(subset=['Statement'],keep='first',inplace=True)
df.info()
df.describe()
```

```python
text = " ".join([x for x in df.Statement])

wordcloud = WordCloud(background_color='white').generate(text)

plt.figure(figsize=(8,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# for neutral
text = " ".join([x for x in df.Statement[df.sentiment=='neutral']])

wordcloud = WordCloud(background_color='white').generate(text)

plt.figure(figsize=(10,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Clustering for dataset exploration

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
iris = load_iris()
samples = iris.data
model = KMeans(n_clusters= 3 ) # kati wata group banaune
model.fit(samples)
labels= model.predict(samples)
```

```python
# plotting sepal length and petal length
xs =samples[:,0]
ys =samples[:,2]
plt.scatter(xs,ys,c=labels)
centroids=model.cluster_centers_
centroids_x=centroids[:,0]
centroids_y=centroids[:,2]
plt.scatter(centroids_x, centroids_y,marker="D",s=50) 
plt.show()
```

**Analysing seeds data form UCI**

```python
seed = pd.read_csv('Seed_Data.csv')
seed.head()
```

```python
seed_dict={
    0:'Kama',
    1:'Ross',
    2:'Canadian',
}
seed_list=seed['target'].map(seed_dict).tolist()
seed_list
```

```python
# Preparing seeds for clustering by dropping the target column
seeds = seed.drop(['target'],axis=1)
seeds.head()
```

```python
ks = range(1,6)
inertias=[] # inertias  mean how much the data are distorted from the centroid of the cluster, less the inertia ,more good 

for k in ks:
    model=KMeans(n_clusters=k)
    model.fit(seeds)
    inertias.append(model.inertia_)

plt.plot(ks,inertias,'-o')
plt.xlabel('Number of Clusters, k')
plt.ylabel('ineritas')
plt.xticks(ks)
plt.show()

# Inertia decreases from 3  to 4 very slowly so, 3 can be a good choice
```

```python
model = KMeans(n_clusters=3)
seed_label = model.fit_predict(seeds)
centroids=model.cluster_centers_
centroids
```

```python
# A vs LKG
xs_A=seeds.iloc[:,0]
ys_LKG=seeds.iloc[:,-1]
centroids_xs_A=centroids[:,0]
centroids_ys_LKG=centroids[:,-1]
plt.scatter(xs_A,ys_LKG,c=seed_label)
plt.scatter(centroids_xs_A,centroids_ys_LKG,marker="D",s=50 ,c='red')
plt.show()
```
